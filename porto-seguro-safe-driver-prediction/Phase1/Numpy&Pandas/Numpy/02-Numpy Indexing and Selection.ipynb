# NumPy 索引&筛选

我们将讨论如何从数组中选择元素或元素组。

```python
import numpy as np
```

```python
arr = np.arange(0,11)
```

```python
arr
```

## 中括号 Bracket Indexing and Selection

选择数组的一个或一些元素的最简单的方法就是下面这样，类似于 python 列表。

```python
arr[8]
```

```python
arr[1:5]
```

```python
arr[0:5]
```

## 广播 Broadcasting

Numpy 数组不同于普通的 Python 列表，它们能够广播：

```python
arr[0:5]=100

arr
```

```python
arr = np.arange(0,11)

arr
```

```python
slice_of_arr = arr[0:6]

slice_of_arr
```

```python
slice_of_arr[:]=99

slice_of_arr
```

注意更改也发生在原始数组中！

```python
arr
```

数据没有被复制，它是原始数组的视图！这样可以避免内存问题！

```python
arr_copy = arr.copy()

arr_copy
```

## 索引二维数组（矩阵）

```python
arr_2d = np.array(([5,10,15],[20,25,30],[35,40,45]))

arr_2d
```

```python
arr_2d[1]
```

```python
arr_2d[1][0] #第二行第一列
```

```python
arr_2d[1,0] #第二行第一列
```

```python
#Shape (2,2) 右上角
arr_2d[:2,1:]
```

```python
arr_2d[2]
```

```python
arr_2d[2,:]
```

### 花式索引

花式索引允许无序选择整个行或列，为了显示这一点，再构建一个 numpy 数组

```python
arr2d = np.zeros((10,10))
```

```python
arr_length = arr2d.shape[1]
```

```python
for i in range(arr_length):
    arr2d[i] = i
    
arr2d
```

Fancy indexing allows the following

```python
arr2d[[2,4,6,8]]
```

```python
# 任何顺序
arr2d[[6,4,2,7]]
```

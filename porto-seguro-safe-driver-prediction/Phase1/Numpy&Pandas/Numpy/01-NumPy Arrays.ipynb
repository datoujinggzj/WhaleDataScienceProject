# Numpy 数组

NumPy 数组是我们在整个课程中使用 Numpy 的主要方式。 Numpy 数组本质上有两种形式：向量和矩阵。

- 向量是严格的一维数组
- 矩阵是二维的（但注意，矩阵仍然只能有一行或一列）。

现在介绍如何创建 NumPy 数组。

## 创建 Arrays

### From a Python List

```python
my_list = [1,2,3]
my_list
```

```python
np.array(my_list)
```

```python
my_matrix = [[1,2,3],[4,5,6],[7,8,9]]
my_matrix
```

```python
np.array(my_matrix)
```

```python
np.array(my_matrix)[:,1]
```

```python
list(zip(*my_matrix))[0]
```

## Built-in 方法

### arange

在给定的间隔内返回均匀间隔的值。

```python
np.arange(0,10)
```

```python
np.arange(0,11,2)
```

### zeros & ones

生成零或一的数组。

```python
np.zeros(3)
```

```python
np.zeros((5,5))
```

```python
np.ones(3)
```

```python
np.ones((3,3))
```

### linspace
Return evenly spaced numbers over a specified interval.

```python
np.linspace(0,10,3)
```

```python
np.linspace(0,10,50)
```

## eye

Creates an identity matrix

```python
np.eye(4)
```

## Random 

产生随机数组

### rand

创建给定形状的数组并填充它
来自均匀分布的随机样本。

```python
np.random.seed(2)
np.random.rand(2)
```

```python
np.random.rand(5,5)
```

### randn

从“标准正态”分布中返回一个（或多个）样本。与均匀分布的`rand`不同

```python
np.random.randn(2)
```

```python
np.random.randn(5,5)
```

### randint

返回从“低”（包括）到“高”（不包括）的随机整数。

```python
np.random.randint(1,100)
```

```python
np.random.randint(1,100,10)
```

## 数组性质（attribute）和方法（method）


```python
arr = np.arange(25)
ranarr = np.random.randint(0,50,10)
```

```python
arr
```

```python
ranarr
```

## 重塑（reshape）

返回包含具有新形状的相同数据的数组。

```python
arr.reshape(5,5)
```

### max / min / argmax / argmin

这些是查找最大值或最小值的方法。或者使用 `argmin` 或 `argmax` 找到它们的索引位置。

```python
ranarr
```

```python
ranarr.max()
```

```python
ranarr.argmax()
```

```python
ranarr.min()
```

```python
ranarr.argmin()
```

## 形状（shape）


形状是数组具有的属性（不是方法）

```python
# Vector
arr.shape
```

```python
# Notice the two sets of brackets
arr.reshape(1,25)
```

```python
arr.reshape(1,25).shape
```

```python
arr.reshape(25,1)
```

```python
arr.reshape(25,1).shape
```
